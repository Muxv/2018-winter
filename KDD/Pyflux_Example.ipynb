```python
import numpy as np
import pandas as pd
import pyflux as pf
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/sunspot.year.csv')
data.index = data['time'].values

plt.figure(figsize=(15,5))
plt.plot(data.index,data['value'])
plt.ylabel('Sunspots')
plt.title('Yearly Sunspot Data');
```

```python
data
```

```python
model = pf.ARIMA(data = data, ar=4, ma=4, target="value", family=pf.Normal())
```

```python
x = model.fit("MLE")
x.summary()
```

```python
model.plot_fit(figsize=(15,5))
```

```python
model.plot_predict(h=100,past_values=10)
```

```python
model.plot_predict_is(h=10,fit_once=False,fit_method='MLE')
```

```python
model.plot_predict_is(h=4,fit_once=True,fit_method='MLE')
```
